```python
Mel_Rental_Market= df[['Suburb','Type', 'Method',
       'Real Estate Agent','Distance', 'Postcode',
       'Number of Bedroom', 'Number of Bathroom', 'number of carpark','Lattitude',
       'Longtitude','Median_rent_weekly']]
```

```python
Mel_Rental_Market=Mel_Rental_Market.rename(columns={'number of carpark':'Number_of_Carpark',
                                                   'Lattitude': 'Latitude',
                                                   'Real Estate Agent': 'Real_Estate_Agent',
                                                   'Number of Bedroom':'Number_of_Bedroom',
                                                   'Number of Bathroom':'Number_of_Bathroom',
                                                   'Longtitude': 'Longitude'})
```

```python
Mel_Rental_Market['id'] = range(len(Mel_Rental_Market))
Mel_Rental_Market.head()
```

```python
#rearrange the columns
Mel_Rental_Market_cleaned=Mel_Rental_Market[['id','Suburb', 'Type', 'Method', 'Real_Estate_Agent', 'Distance', 'Postcode',
       'Number_of_Bedroom', 'Number_of_Bathroom', 'Number_of_Carpark',
       'Latitude', 'Longitude', 'Median_rent_weekly']]
```

```python
Mel_Rental_Market_cleaned.head()
```

```python
conn = sqlite3.connect('Melbourne.db')

# Save the DataFrame to an SQLite table
Mel_Rental_Market_cleaned.to_sql('Melbourne', conn, if_exists='replace', index=False)

# Close the database connection
conn.close()
```

```python
# Connect to the database
conn = sqlite3.connect('Melbourne.db')
cursor = conn.cursor()

# Load the data into a DataFrame
df = pd.read_sql_query("SELECT * FROM Melbourne", conn)

# Create a new table with a primary key constraint
cursor.execute("CREATE TABLE Melbourne_2 (id INTEGER PRIMARY KEY, Suburb TEXT, Type TEXT, Method TEXT, Real_Estate_Agent TEXT, Distance REAL, Postcode INTEGER, Number_of_Bedroom INTEGER, Number_of_Bathroom INTEGER, Number_of_Carpark INTEGER, Latitude REAL, Longitude REAL, Median_rent_weekly INTEGER)")

# Commit the changes and close the connection
conn.commit()
conn.close()

```
